```python
import pandas as pd
teleconX = pd.read_json('/content/TelecomX_Data.json')
teleconX
```

```python
teleconX_customer = pd.json_normalize(teleconX['customer'])
teleconX_customer
```

```python
teleconX_phone = pd.json_normalize(teleconX['phone'])
teleconX_phone
```

```python
teleconX_internet = pd.json_normalize(teleconX['internet'])
teleconX_internet
```

```python
teleconX_account = pd.json_normalize(teleconX['account'])
teleconX_account
```

```python
dados = pd.concat([teleconX, teleconX_customer, teleconX_phone, teleconX_internet, teleconX_account], axis=1)
dados = dados.drop(columns=['customer', 'phone', 'internet', 'account'])
dados.head()
```

#ðŸ”§ TransformaÃ§Ã£o

```python
dados.info()
```

```python
import  numpy as np
dados['Charges.Total'] = dados['Charges.Total'].replace(' ', np.nan, regex=True)
dados['Charges.Total'] = dados['Charges.Total'].replace('', np.nan, regex=True)
dados['Charges.Total'] = dados['Charges.Total'].astype(np.float64)
dados.info()
```

```python
dados['Churn'].unique()
```

```python
dados['Churn'] = dados['Churn'].replace('', np.nan, regex=True)
```

```python
dados.dropna(subset=['Charges.Total'], inplace=True)
dados.dropna(subset=['Churn'], inplace=True)
dados.info()
```

```python
dias_no_mes = 30
dados['Contas_Diarias'] = dados['Charges.Monthly'] / dias_no_mes
print("DataFrame com a nova coluna 'Contas_Diarias':")
print(dados[['Charges.Monthly', 'Contas_Diarias']].head())
```
